**下面给出两个常见的注意力评分函数**

   
**当q，k维度不同时（加性注意力）：**  
\begin{equation}
a(q, k) = w^\top \tanh(W_q q + W_k k)
\end{equation}
   
**当q，k维度相同时（缩放点积注意力）：**   
\begin{equation}
a(q, k) = \frac{q^\top k}{\sqrt{d}}
\end{equation}

\begin{equation}
\text{Attention}(Q, K, V) = \text{softmax} \left( \frac{Q K^\top}{\sqrt{d}} \right) V
\end{equation}

**位置编码**

**矩阵第i行第2j列和2j+1列所添加的位置信息分别为：**   

\begin{align}
PE_{(i, 2j)} &= \sin\left( \frac{i}{10000^{\frac{2j}{d}}} \right) \\
PE_{(i, 2j+1)} &= \cos\left( \frac{i}{10000^{\frac{2j}{d}}} \right)
\end{align}
